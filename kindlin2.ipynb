```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cancer_types_summary (1).txt', sep='\t')
df
```

```python
pivot_df = df.pivot(index='Cancer Study', columns='Alteration Type', values='Alteration Frequency').reset_index()

# Rename the columns for clarity
pivot_df.columns.name = None  # Remove the columns' name
pivot_df = pivot_df.rename(columns={'Cancer Study': 'Cancer Study'})

# Fill NaN values with 0 if needed
pivot_df = pivot_df.fillna(0)

# Display the resulting DataFrame
pivot_df =pivot_df.set_index('Cancer Study')
pivot_df['sum'] = pivot_df.sum(axis=1)
pivot_df = pivot_df.sort_values('sum', ascending=False)
pivot_df = pivot_df.iloc[:, :-1]
pivot_df
```

```python
custom_colors = ['midnightblue', 'blueviolet', 'magenta', 'deeppink', 'darkturquoise',
                'mediumspringgreen', 'olivedrab', 'coral', 'firebrick']

fig, ax = plt.subplots(figsize=(10, 4))
pivot_df.plot(kind='bar', stacked=True, ax=ax, color=custom_colors)
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
ax.tick_params(axis='both', labelsize=12)
plt.savefig('stack_freq.png', dpi=600, bbox_inches='tight')
```

```python
df = pd.read_csv('PATIENT_DATA_oncoprint.tsv', sep='\t', index_col=(0)).T
df = df[~df.index.duplicated(keep='first')].T
# df = df[df.index.isin(['FERMT3_CNA', 'FERMT3_MUTATIONS', 'FERMT3_MRNA', 'FERMT3_PROTEIN', 'FERMT3_SV'])]
df
```

```python
types = []
for col in df:
    if df[col].isna().all():
        types.append('Unaltered')
    else:
        types.append('Altered')
len(types)

data = pd.DataFrame([df.columns, types]).T
data.to_csv('Fermt3_types.csv', index=False)
```
